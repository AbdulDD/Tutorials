# Groupby

```python
DF.groupby("Item ID")["Price(Rs)"].min()
DF.groupby("Item ID")["Price(Rs)"].max()
DF.groupby("Item ID")["Price(Rs)"].mean()
DF.groupby("Item ID")["Price(Rs)"].count()
DF.groupby("Item ID")["Price(Rs)"].sum()
```

# Map

```python
Price_count_with_Item_ID = DF.groupby("Item ID")["Price(Rs)"].count()
Price_count_with_Item_ID

DF["Pricecount_respectiveto_ItemID"] = DF["Item ID"].map(Price_count_with_Item_ID)
```

# Transform

```python
DF["Pricecount_respectiveto_ItemID_T"] = DF.groupby("Item ID")["Price(Rs)"].transform('count')
DF["Weight"] = DF.groupby("Item ID")["Weight"].transform('sum')

# Not to work on categorical data
#DF["Touch"] = DF.groupby("Item ID")[" Touch "].transform('')
```
